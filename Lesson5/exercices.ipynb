```python
import pandas as pd
import numpy as np
```

```python
clean_people = pd.read_csv('../Lesson4/data/clean_people.csv')
```

```python
clean_people.head()
```

```python
# Set inactive people
current_date = pd.datetime.now()
clean_people['last_seen'] = pd.to_datetime(clean_people['last_seen'])
clean_people.loc[current_date - clean_people['last_seen'] >= np.timedelta64(1, 'Y'), 'inactive'] = True
```

```python
# Filter phone number
valid_number = clean_people.loc[clean_people.phone.str.contains(r"^0[0-9]{9}").fillna(False)].copy()
valid_number.head()
```

```python
# Get mobile phone
valid_number['is_mobile_phone'] = 0
mask_phone_number = (valid_number.phone.str.startswith('06')) | (valid_number.phone.str.startswith('07'))
valid_number.loc[mask_phone_number, 'is_mobile_phone'] = 1
```
